## Feature Engineering

주당_카페_방문횟수, 포인트_쿠폰_관리, 적립_빈도 컬럼은 객관식 선택형으로 최대 5개정도의 범주로 보아 레이블인코딩이 가능해보인다.  
하지만 카페_선택_이유 컬럼의 경우 다지선다형으로 원핫인코딩으로 각각의 선택지를 더미변수로 전환해 해결해야한다.

```python
# 카페_선택_이유 컬럼 변환
df = df.join(df['카페_선택_이유'].str.get_dummies(sep=";")).drop(columns=['카페_선택_이유'])
```

```python
# Likert Scale 매핑
visit_likert_mapping = {
    "1회 이상":1,
    "2회 이상":2,
    "3회 이상":3,
    "4회 이상":4,
    "매일":5,
}

# 변환 적용
df["주당_카페_방문횟수"] = df["주당_카페_방문횟수"].map(visit_likert_mapping)
```

```python
# Likert Scale 매핑
point_likert_mapping = {
    "매우 그렇다": 4,
    "그렇다": 3,
    "보통": 2,
    "그렇지 않다": 1,
    "매우 그렇지 않다": 0,
}

# 변환 적용
df["포인트_쿠폰_관리"] = df["포인트_쿠폰_관리"].map(point_likert_mapping)
```

```python
# Likert Scale 매핑
accumulate_likert_mapping = {
    "매번 한다": 7,
    "가끔 한두번은 하지 않는 것 같다": 5,
    "할 때도 있고 안 할 때도 있다": 3,
    "거의 하지 않는다": 1,
    "아예 하지 않는다": 0,
}

# 변환 적용
df["적립_빈도"] = df["적립_빈도"].map(accumulate_likert_mapping)
```
