```python
# ✅ 3. '카페_선택_이유' 컬럼을 원핫 인코딩 (복수 응답 가능 범주형)
df = df.join(df["카페_선택_이유"].str.get_dummies(sep=";")).drop(columns=["카페_선택_이유"])
```

```python
# ✅ 4. '포인트_쿠폰_관리'와 'Q3'는 Factorizing (라벨 인코딩)
df["주당_카페_방문횟수"] = df["주당_카페_방문횟수"].astype("category").cat.codes
df["포인트_쿠폰_관리"] = df["포인트_쿠폰_관리"].astype("category").cat.codes
df["적립_빈도"] = df["적립_빈도"].astype("category").cat.codes

print(df)
```
